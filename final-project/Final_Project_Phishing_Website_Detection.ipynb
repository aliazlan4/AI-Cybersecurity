This is a Machine Learning based phishing website detector. It is based on Keras and Tensorflow ML model.

```
# lets remove the warnings
import warnings
warnings.filterwarnings("ignore")

# import all required libraries
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import pandas as pd
import numpy as np
```

Fetching dataset from CSV file and then updating it a little bit for the training

```
data = pd.read_csv('phishing-dataset.csv')

# updating dataset a little bit for keras
del data['id']
data['Result'] = data['Result'].map({-1:0, 1:1})

data.head(10)
```

Splitting the dataset into 75% and 25% sets for training and testing purpose.

```
y = data.iloc[:, -1].values
x = data.iloc[:, :-1].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=np.random.seed())
```

We are creating a Keras Sequential model here with a learning rate of 0.01.

```
model = tf.keras.Sequential()

model.add(layers.InputLayer(input_shape=(1,30)))
model.add(layers.Dense(100, activation='relu'))
model.add(layers.Dense(50, activation='relu'))
model.add(layers.Dense(1,  activation='sigmoid'))

model.compile(tf.keras.optimizers.Adam(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

Training and evaluating our model.

```
callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=0.001, patience=10)

history = model.fit(x_train, y_train, batch_size=64, epochs=128, validation_split=.2, callbacks=[callback])

scores = model.evaluate(x_test, y_test)
```

```
print('\nAccuracy score of the Neural Network with basic hyperparameter settings {0:.2f}%'.format(scores[1]*100))
```

Saving our model as a file to use it afterwards for testing new websites.

```
model.save('model.h5') 

print("model saved.")
```
